```python
#first run part 1 and then 2, then 3
#QUESTION 4 PART 1

import pandas as pd
from openpyxl import load_workbook

sheet_names = ["Monday", "Tuesday", "Wednesday", "Thursday"]
sws = []

for sheet_name in sheet_names:
    # Load the Excel file and sheet
    excel_file = pd.ExcelFile("TT.xlsx")
    data = excel_file.parse(sheet_name, skiprows=4)

    # Load the sheet with openpyxl to access cell colors
    wb = load_workbook("TT.xlsx", data_only=True)
    ws = wb[sheet_name]

    # Create a list of dictionaries to store data and colors
    data_with_colors = []

    for row in ws.iter_rows():
        row_data = []
        for cell in row:
            # Extract cell color
            fill = cell.fill
            cell_color = fill.start_color.index if fill.start_color else None
            row_data.append([cell.value,cell_color])
        data_with_colors.append(row_data)

    # Convert the list of dictionaries to a DataFrame
    df = pd.DataFrame(data_with_colors)

    # Append the DataFrame to the list
    sws.append(df)

```

```python

#QUESTION 4 PART 2


batch_name = ["BS CS (2020)","BS CS (2021)","BS CS (2022)","BS CS (2023)",
         "BS DS (2020)","BS DS (2021)","BS DS (2022)","BS DS (2023)",
         "BS AI (2020)","BS AI (2021)","BS AI (2022)","BS AI (2023)",
         "BS CY (2020)","BS CY (2021)","BS CY (2022)","BS CY (2023)",
         "BS SE (2020)","BS SE (2021)","BS SE (2022)","BS SE (2023)"]
color_name = ["FFFFE599","FFC39401","FF6D5200","FFFFB740",
         "FFB4A7D6","FFB17FD7","FF351C75","FF7F4CFF",
         "FFB6D7A8","FF6AA84F","FF274E13","FF00F600",
         "FFABCCEB","FF599DDA","FF073763","FF0000FF",
         "FFF4CCCC","FFDD7E6B","FF85200C","FFE62C06"]
rmcl = "FFA5A5A5"
batch = input("Enter batch name : ")

# take color of batch
for i in range(20):
    if batch_name[i] == batch:
        color = color_name[i]
for i in range(4):
    sheet = sws[i]
    col = sheet.shape[0]-1
    row = sheet.shape[1]-1
    for j in range(col):
        for k in range(row):
            #removing color from other batches
            clvl = sws[i][k][j]
            if color in clvl:
                sws[i][k][j] = clvl[0]
            elif rmcl in clvl:
                sws[i][k][j] = clvl[0]
            else:
                sws[i][k][j] = " "
                
                
            
        
```

```python

#QUESTION 4 PART 3


import pandas as pd
from openpyxl import Workbook
from openpyxl.styles import PatternFill
output_file = "output_with_colors.xlsx"
wb = Workbook()
for i in range(4):
    data = sws[i]
    sheet_name = sheet_names[i]
    ws = wb.create_sheet(title=sheet_name)
    col = data.shape[0]-1
    row = data.shape[1]-1
    for j in range(col):
        for k in range(row):
            #filling value and color with cell
            ws.cell(row=j+1, column=k+1, value=data[k][j])
            if data[k][j] == " ":
                continue
                fill = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")
            else:
                fill = PatternFill(start_color=color, end_color=color, fill_type="solid")
            ws.cell(row=j+1, column=k+1).fill = fill
if 'Sheet' in wb.sheetnames:
    wb.remove(wb['Sheet'])
wb.save(output_file)
print(f"Saving the Excel file with colors to {output_file}")
```
